```
import numpy as np 
import plotly
import plotly.graph_objects as go
import plotly.express as px
import random as rd
import plotly.offline as py
```

```
fig = go.Figure(data=go.Bar(y=[1,2,3,4,2,1]))
fig.show()
```

```
x = np.linspace(0,10)
y = x**2
yerr = 3*np.random.rand(50)
px.line(x = x,y=y)
px.scatter(x=x,y=y,error_y = yerr,labels={"x":"X Achse","y":"y-Achse"},hover_name=np.arange(0,len(y),1))
```

```
def f(t):
    return np.cos(t),np.sin(t),5*t
t= np.linspace(0,20,1000)
fx,fy,fz = f(t)
fig = px.line_3d(None,fx,fy,fz)
fig.update_traces(line_color='#ff3439')
```
